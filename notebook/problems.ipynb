We will use NumPy to obtain information to describe statistically.

- Generate an array of 100 elements following a normal distribution.
- Generate an array of 100 elements following a chi-square distribution with 3 degrees of freedom.
- Calculate the main metrics and statistical measures that best describe the two vectors.

````python
# TODO


import numpy as np

import statistics as stats

# Generate an array of 100 elements following a normal distribution

normal_dist = np.random.normal(size=100)

# Generate an array of 100 elements following a chi-square distribution with 3 degrees of freedom
chi2_dist = np.random.chisquare(3, size=100)

print(f"Vector Normal: {normal_dist}")
print(f"Chi_Square Vector: {chi2_dist}")


# Media:

import statistics as stats

media_normal = np.mean(normal_dist)
print(f'Mean (average): {media_normal:.2f}')

# Desviaci贸n Standar:

std_normal = np.std(normal_dist)
print(f'Standard Desviaci贸n: {std_normal:.2f}')

# Varianza:

varianza_normal = np.var(normal_dist)
print(f'Variance: {varianza_normal:.2f}')

# Skewness:

from scipy.stats import skew

skew_normal = skew(normal_dist)


print(f"Normal skewness: {normal_dist}")


# Kurtosis Normal:

from scipy.stats import kurtosis

kurt_normal = kurtosis(normal_dist)


print(f"Kurtosis Normal: {kurt_normal}")

# Datos de Chi:

#Media:

media_chi2 = np.mean(chi2_dist)
print(f'Mean (average): {media_chi2:.2f}')

# Varianza de Chi:

variance_chi2 = np.var(chi2_dist)
print(f'Variance: {variance_chi2:.2f}')

# Moda de Chi:

mode_chi2 = stats.mode(chi2_dist)
print(f'Mode: {mode_chi2}')

# Mediana de Chi:

median_chi2 = np.median(chi2_dist)
print(f'Median: {median_chi2:.2f}')

# Desviaci贸n standar de chi:

std_chi2 = np.std(chi2_dist)
print(f'Standard Deviation: {std_chi2:.2f}')

# Skewness:

from scipy.stats import skew

skewness_chi2 = skew(chi2_dist)
print(f'Skewness: {skewness_chi2:.2f}')

# Kurtosis Normal:

from scipy.stats import kurtosis

kurtosis_chi2 =kurtosis(chi2_dist)
print(f'Kurtosis: {kurtosis_chi2:.2f}')



````
